Ques -2 Find the most effective tweet (create a measure of your own based on parameters such as
#retweets, time from death, etc.)


```python
# here i have calculated the effective of the tweet by the retweetscount/ time 
```

```python

df['effectiveness'] = df['retweets_count'] / df['time'].astype(int)


most_effective_tweet_index = df['effectiveness'].idxmax()


most_effective_tweet = df.loc[most_effective_tweet_index, 'tweet']
most_effective_time = df.loc[most_effective_tweet_index, 'time']
most_effective_retweets = df.loc[most_effective_tweet_index, 'retweets_count']

print(f"Most effective tweet: {most_effective_tweet}")
print(f"Time: {most_effective_time}")
print(f"Retweets count: {most_effective_retweets}")
```
